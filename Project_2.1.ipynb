```python
import pandas as pd # Nos permite manejar data como dataframes
import numpy as np # Numpy es una librería de programación numérica super rápida
import matplotlib.pyplot as plt #sets up plotting under plt
SL = pd.read_csv('synergy_logistics_database.csv')
SL
```

```python
cyw1= ['register_id','direction', 'origin', 'destination', 'total_value']
group1 = SL[cyw1].groupby( by= ['direction', 'origin', 'destination'])
group1[cyw1]
```

```python
group_dem= group1['register_id'].count()
group_dem
```

```python
order_dem = group_dem.sort_values(ascending=False)#.head()
order_dem
```

```python
dem_df= order_dem.to_frame()
dem_df
```

```python
rn = dem_df.rename(columns={'register_id':'Demand'})
rn.head(10)
```

```python
group_rev=group1['total_value'].sum()
group_rev
```

```python
routes= []
list_values1 = []
list_demand = []
for index, value in group_rev.items():
    for ind, val in order_dem.items():
        if index == ind:
            print(f"Route: {index}, Revenue: {value}, Demand: {val} ")
            routes.append(index)
            list_values1.append(value)
            list_demand.append(val)
```

```python
df1 = pd.DataFrame(list(zip( routes,list_values1,list_demand)), 
               columns =['Route', 'Total_Revenue', 'Demand' ]) 
Top_10 =df1.sort_values(by='Demand', ascending=False).head(10)
Top_10
```

```python
Total_Rev_Dem=Top_10['Total_Revenue'].sum()
Total_Rev_Dem
total_rev = SL['total_value'].sum()
total_rev
Per_Top_10 = Total_Rev_Dem/total_rev * 100
Per_Top_10
print ('Total Revenue:','$',Total_Rev_Dem, '\nRevenue_Top10_Routes:','$',total_rev, '(',Per_Top_10, '%) of Total'  )
```
